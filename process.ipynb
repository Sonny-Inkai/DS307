##  Tao users signin 

```python
import pandas as pd
import json

# Đọc file Excel
df = pd.read_csv('hotels_users_ratings.csv')
```

```python
users_data = df[['UserID', 'User']].drop_duplicates()

# Chuyển đổi dữ liệu thành danh sách các từ điển
users_list = users_data.to_dict(orient='records')

# Xuất danh sách thành file JSON
with open('users.json', 'w', encoding='utf-8') as json_file:
    json.dump(users_list, json_file, ensure_ascii=False, indent=4)

print("Đã tạo file users.json thành công!")
```

```python
df[df['HotelID'] ==  4]['Rating'].values.mean()
```

```python
hotel_ratings = []
for i in hotel_ids:
    rating = df[df['HotelID'] == i]['Rating'].mean()
    scaled_rating = rating / 2
    
    rating = round(scaled_rating * 2) / 2
    
    hotel_ratings.append(rating)
```

```python
hotel_df = df.drop_duplicates(subset=['HotelID'])
```

```python
hotel_df['Rating'] = hotel_ratings
```

```python
hotel_df
```

```python
image_paths = [f"/img/hotel/hotel{i}.webp" for i in range(1, 16)]
```

```python
import random

hotel_df["images"] = hotel_df.apply(lambda _: random.sample(image_paths, k=4), axis=1)
```

```python
hotel_df
```

```python
hotels_json = []
for _, row in hotel_df.iterrows():
    hotel = {
        "id": row["HotelID"],
        "name": row["Name Hotel"],
        "location": row["Location"],
        "description": row["Descriptions"],
        "address": row["Address"],
        "rating": row["Rating"],
        "images": [{"image1": row["images"][0]}, {"image2": row["images"][1]}, {"image3": row["images"][2]}, {"image3": row["images"][3]}]
    }
    hotels_json.append(hotel)

with open("hotels.json", "w", encoding="utf-8") as f:
    json.dump(hotels_json, f, ensure_ascii=False, indent=4)

```
