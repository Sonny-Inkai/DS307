```python
from mplfonts import use_font
use_font('Noto Serif CJK SC')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu
df = pd.read_csv('hotels_users_ratings.csv')

# 1.1 Kiểm tra thông tin cơ bản
print("Thông tin cơ bản:")
print(df.info())
print("\nThống kê mô tả:")
print(df.describe())

# 1.2 Kiểm tra missing values
print("\nMissing values:")
print(df.isnull().sum())

# 1.3 Phân tích phân phối rating
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Rating', bins=10)
plt.title('Phân phối Rating')
plt.show()

# 1.4 Phân tích theo location
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Location', y='Rating')
plt.xticks(rotation=45)
plt.title('Rating theo Location')
plt.show()
```
